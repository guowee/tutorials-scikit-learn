## 数据清洗
#### 格式转化
   credits数据中，cast、crew都是json的格式，需要将演员、导演读取出来，以字符串格式显示<p>
   movies数据中genres、keywords、production_companies、spoken_languages也是json格式，需要转化成字符串
* credits json 解析

```python
json_cols = ['cast','crew']
# 应用json.loads
# 编码： 把一个Python对象编码转换成Json字符串 json.dumps()
# 解码： 把Json格式字符串解码转换成Python对象 json.loads()
for i in json_cols:
    credits[i]=credits[i].apply(json.loads)
#用json方法汇总所有类型
def get_names(x):
    return ','.join([i['name'] for i in x])
# i['name']提取name后的名字，以逗号分隔
# json() 连接字符串数组，将字符串，元组，列表中的元素以指定的字符（分隔符）连接生成一个新的字符串
credits['cast']=credits['cast'].apply(get_names)
credits.head()
```

```python
#crew 提取导演director
def director(x):
    for i in x:
        if i['job']=='Director':
            return i['name']
credits['crew'] = credits['crew'].apply(director)
credits.head()
```

```python
#将字段 'crew’改为‘director’
credits.rename(columns ={'crew':'director'},inplace=True)
credits.head()
```

* movies json解析

```python
json_cols = ['genres','keywords','spoken_languages','production_countries','production_companies']
for i in json_cols:
    movies[i]=movies[i].apply(json.loads)
def get_names(x):
    return ','.join([i['name'] for i in x])
movies['genres']=movies['genres'].apply(get_names)
movies['keywords']=movies['keywords'].apply(get_names)
movies['spoken_languages']=movies['spoken_languages'].apply(get_names)
movies['production_countries']=movies['production_countries'].apply(get_names)
movies['production_companies']=movies['production_companies'].apply(get_names)

movies.head()
```

## 合并数据
credits和movies中都有movie_id和title,检查这两个字段是否是相同的

```python
(credits['movie_id']==movies['id']).describe()
```

```python
(credits['title']==movies['title']).describe()
```

可以看出这两个字段是相同的，将movies的title删掉

```python
del movies['title']
df = credits.merge(right=movies,how='inner',left_on='movie_id',right_on='id')
df.head()
```

## 字段及缺省值处理 

```python
# 删除不必要字段
del df['id']
del df['homepage']
del df['original_title']
del df['overview']
del df['spoken_languages']
del df['tagline']
df.head()
```

```python
df.info()
```

> 可以看出director,release_date,runtime分别有30,1,2个缺省值，这里只处理release_date,runtime的缺省值

```python
df['release_date']=df['release_date'].fillna('2014-06-01')
df['runtime']=df['runtime'].fillna(df.runtime.mean())
df[df[['release_date','runtime']].isnull().values ==True]
df.head()
```

# 数据可视化分析
## 关于电影类型的分析
* 获取电影类型

```python
df['genres'][1].split(',')
```

```python
#set() 函数创建一个无序不重复元素集，可进行关系测试，删除重复数据，还可以计算交集，并集，差集等
#创建一个set（）
#创建一个循环
#union(联合)取并集set(a).union(b)
genre=set()
for i in df['genres'].str.split(','):
    genre=set().union(i,genre)
genre = list(genre)
genre.remove('')
genre
```

可以看出电影类型主要有这20种
* 转化日期格式


```python
df['release_year']=pd.to_datetime(df.release_date,format='%Y-%m-%d').dt.year
df['release_month']=pd.to_datetime(df.release_date,format='%Y-%m-%d').dt.month
df.head()
```

## 电影类型与数量的关系
* 分类型判断每部电影属于的类型

```python
for genr in genre:
    df[genr]=df['genres'].str.contains(genr).apply(lambda x:1 if x else 0)
df_gy = df.loc[:,genre]
df_gy.head()
```

* 建立包含电影类型和年份的数据框

```python
df_gy.index = df['release_year']
df_gy.head()
```

```python
#电影数量的逐年趋势变化
df_gy.groupby('release_year').count().Adventure.plot(figsize=(18,9))
plt.xticks(range(1915,2018,5))
plt.title('电影数量的年度趋势图',fontsize=20)
plt.xlabel('年份',fontsize=18)
plt.ylabel('数量',fontsize=18)
plt.show()
```

> 可以看出电影的拍摄数量从1992年开始大幅增加

```python
df_gy1=df_gy.sort_index(ascending=False) #根据索引年份排序
df_gy1.tail()
```

* 统计每个类型每年的电影数量

```python
df_gys = df_gy1.groupby('release_year').sum()
df_gys.tail()
```

* 电影类型随时间变化的趋势图（折线图）

```python
plt.figure(figsize=(18,9))
plt.plot(df_gys,label=df_gys.columns)
plt.legend(df_gys)
plt.xticks(range(1915,2018,5))
plt.title('电影类型随时间的变化',fontsize=20)
plt.xlabel('年份',fontsize=18)
plt.ylabel('数量',fontsize=18)
plt.show()
```

可以看出Drama和War这两个类型是随时间波动比较大的，增长较快，现在仍比较热门<br>

电影类型的数量关系（各个类型的总数量）

```python
df_gysum=df_gys.sum().sort_values(ascending=True)
df_gysum
```

* 各个电影类型的数量条形图

```python
df_gysum.plot.barh(label='genre',figsize=(18,9))
plt.title('电影类型的数量条形图',fontsize=20)
plt.xlabel('数量',fontsize=18)
plt.ylabel('电影类型',fontsize=18)
plt.show()
```

> 可以看出Drama和Comedy是数量最多的前两个<br>

各个电影类型的比例饼图

```python
bl = df_gysum/df_gysum.sum()
explode=(bl>=0.06)/20+0.02
df_gysum.plot.pie(autopct='%1.1f%%',figsize=(12,12),label='',explode=explode)
plt.title('电影类型比例分布图',fontsize=20)
plt.show()
```

> Drama、Comedy、Thriller、Action这四个类型占了一半多比例，分别为18.9%，14.2%，10.5%，9.5%

## 电影类型与利润的关系

### 电影类型与利润的关系

```python
#定义利润
df['profit']=df['revenue']-df['budget']
df_gen_pro=pd.DataFrame(index=genre)
#求出每种类型的平均利润
list=[]
for genr in genre:
    list.append(df.groupby(genr)['profit'].mean())
list2=[]
for i in range(len(genre)):
    list2.append(list[i][1])
df_gen_pro['mean_profit']=list2
df_gen_pro.head()
```

电影类型的利润条形图

```python
df_gen_pro.sort_values(by='mean_profit',ascending=True).plot.barh(label='genre',figsize=(16,8))
plt.title('电影类型的利润条形图',fontsize=20)
plt.xlabel('利润',fontsize=18)
plt.ylabel('电影类型',fontsize=18)
plt.grid(True)
plt.show()
```

可以看出Animation和Advanture,Fantasy这三个类型的电影是盈利最好的，而Foreign和TV Movie这两个是会亏损的
